```r
pgbench <- read.table("./m5d_48vms_pgbench_no_idle_11_11_2019 - Data.csv", sep=",", header=TRUE)
head(pgbench)
```

```r
temp_vm_id <- subset(pgbench, subset=instanceID!="")
vm_id <- factor(temp_vm_id$instanceID)

#table(vm_id)
vm_ids <- as.integer(vm_id)
print(min(vm_ids))
print(max(vm_ids))
levels(vm_id)
```

```r
#Try to map each experiment id to a vm id 1_48

#Get rid of the empty string that snuck into our data
temp_vm_id <- subset(pgbench, subset=instanceID!="")

#Create a factor of instanceIDs and then map them to integers.  
#These will be our vmIDs
vm_id <- factor(temp_vm_id$instanceID)

#table(vm_id)
vm_ids <- as.integer(vm_id)
print(min(vm_ids))
print(max(vm_ids))
levels(vm_id)
```

```r
#Now bind the list of vm ids back into our dataframe
#pgbench <- cbind(pgbench, exp_ids)
pgbench <- pgbench[pgbench$instanceID != "",]
pgbench <- cbind(pgbench, vm_ids)
```

```r
tail(pgbench)
```

```r
#Accidentially add some useless columns
drop <- c("X", "X.1", "X.2", "X.3", "X.4", "vm_ids.1")
pgbench <- pgbench[, !colnames(pgbench) %in% drop]
head(pgbench)
```

```r
#We have some data that was intentionally removed because some transactions got bursty
#We should replace this data with the max found in this column.
#pgbench[pgbench$transactions == "NA", ]
trans_max <- max(pgbench$transactions, na.rm=TRUE)
pgbench$transactions[is.na(pgbench$transactions)] <- trans_max
#tail had been showing some of the NA values for transactions
tail(pgbench)
```

```r
#Looks like there is no missing data.
print(pgbench[pgbench$transactions == "NA", ])
print(pgbench[pgbench$setid == "NA", ])
print(pgbench[pgbench$vm_ids == "NA", ])
```

```r
#There are exactly as many observations as there should be: ((48 * (48 + 1)) / 2) * 3 = 3528
dim(pgbench)
head(pgbench)
```

```r
pgbench <- pgbench[c("transactions","setId","vm_ids")]
colnames(pgbench) <-c("pgbench","setId","vmId")
write.csv(pgbench,"./pgbench_standardized.csv",row.names = FALSE)
```
