```r
library(dplyr)
data<-read.csv("./merged.csv")
#data range differnet between columns won't influence linear model
#cause we can easily do linear transformation 
```

```r
modelLinear = lm(data$set ~ data$iperf + data$sysbench + data$ycruncher + data$pgbench)
summary(modelLinear) 
```

```r
predicted <- predict(modelLinear, data)
cat("R-squared:",cor(predicted, data[,'set'])^2,"\n")
```

```r
par(mfrow=c(2,2))
plot(modelLinear)
 
```
