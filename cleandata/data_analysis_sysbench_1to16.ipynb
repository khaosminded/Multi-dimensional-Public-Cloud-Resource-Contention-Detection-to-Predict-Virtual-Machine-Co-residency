# Import data

```r
library(dplyr)
library("ggpubr")
samples<-read.csv("./named/sysbench_all//sysbench_c4.large_10run_1to16_labled.csv")
```

```r
samples["total.time"] = as.numeric(gsub("s","",unlist(samples["total.time"])))
samples
```

```r
statistic_result=data.frame()

for(i in 1:16){
    thisSet=samples[samples$set==i,'total.time']
    statistic_result=rbind(statistic_result,c(range(thisSet),mean(thisSet),median(thisSet),var(thisSet),i))
}
names(statistic_result)<-c('min','max','mean','median','var','set')
statistic_result

ggplot(statistic_result, aes(x = set, y = mean, colour = set)) +
geom_line() + scale_y_discrete(limits = c(8.45,8.46))
```

```r
ggboxplot(samples, x = "set", y = "total.time", 
          color = "set", 
          ylab = "seconds", xlab = "set",facet.by='vmID')
```
